```python
#This will help you understand the distribution of numerical features like age, bmi, HbA1c_level, and blood_glucose_level.
import matplotlib.pyplot as plt
numerical_columns = ['age','bmi','HbA1c_level','blood_glucose_level']
model[numerical_columns].hist(figsize=(10,8), bins=20)
plt.tight_layout()
plt.show()
```

```python
#boxplot to detect outliers
import seaborn as sns
plt.figure(figsize=(12,8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2,2,i)
    sns.boxplot(model[column])
    plt.title(f'Boxplot of {column}')
    plt.tight_layout()
    plt.show()
```

```python
# IQR (Interquartile Range):
#Use the IQR method to detect outliers. Points outside the range of 1.5 times the IQR from the quartiles are considered outliers.
Q1 = model[numerical_columns].quantile(0.25)
Q3 = model[numerical_columns].quantile(0.75)
IQR = Q3 - Q1
outliers = (model[numerical_columns] < (Q1 - 1.5 * IQR)) | (model[numerical_columns] > (Q1 + 1.5 * IQR))
print(f'Number of outliers:\n{outliers.sum()}')
```

```python
#HANDLING OUTLIERS
#removing outliers based n IQR
model_cleaned = model[~outliers.any(axis=1)]
#checking the shape of the cleaned dataset
print(f"Shape of dataset after removing outliers: {model_cleaned.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Define the models
model1 = LogisticRegression(class_weight='balanced', random_state=42)
model2 = RandomForestClassifier(class_weight='balanced', random_state=42)

# Define the pipeline with scaling and voting classifier
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Standardize features
    ('classifier', VotingClassifier(estimators=[
        ('lr', model1),
        ('rf', model2)
    ], voting='soft'))
])

# Fit the pipeline with training data
pipeline.fit(X_train, y_train)

def predict_diabetes(patient_data):
    """
    Predicts whether a patient has diabetes or not.
    
    Parameters:
    patient_data (list): Feature values of the patient for prediction.
    
    Returns:
    str: Prediction result ('Diabetic' or 'Non-Diabetic').
    """
    # Ensure patient_data is in the correct format (2D array for a single instance)
    patient_data = [patient_data]  # Convert to a 2D array

    # Make prediction using the pipeline
    prediction = pipeline.predict(patient_data)

    # Map the prediction to a human-readable label
    if prediction[0] == 1:
        return 'Diabetic'
    else:
        return 'Non-Diabetic'

def get_user_input():
    """
    Prompts the user for their medical information and returns it in the format required by the model.
    """
    # Get user input
    age = float(input("Enter age: "))
    hypertension = int(input("Enter hypertension (0 or 1): "))
    heart_disease = int(input("Enter heart disease (0 or 1): "))
    bmi = float(input("Enter BMI: "))
    HbA1c_level = float(input("Enter HbA1c level: "))
    blood_glucose_level = float(input("Enter blood glucose level: "))
    
    # For simplicity, assume gender is Male (1.0), Female (0.0), Other (0.0)
    gender = input("Enter gender (Male/Female/Other): ").strip().lower()
    gender_Female = 1.0 if gender == 'female' else 0.0
    gender_Male = 1.0 if gender == 'male' else 0.0
    gender_Other = 1.0 if gender == 'other' else 0.0
    
    # Assume smoking history (No Info, Current, Ever, Former, Never, Not Current)
    smoking_history = input("Enter smoking history (No Info/Current/Ever/Former/Never/Not Current): ").strip().lower()
    smoking_history_No_Info = 1.0 if smoking_history == 'no info' else 0.0
    smoking_history_current = 1.0 if smoking_history == 'current' else 0.0
    smoking_history_ever = 1.0 if smoking_history == 'ever' else 0.0
    smoking_history_former = 1.0 if smoking_history == 'former' else 0.0
    smoking_history_never = 1.0 if smoking_history == 'never' else 0.0
    smoking_history_not_current = 1.0 if smoking_history == 'not current' else 0.0
    
    health_index = float(input("Enter health index: "))
    log_blood_glucose = float(input("Enter log blood glucose: "))
    
    # Prepare the patient data
    patient_data = [
        age, hypertension, heart_disease, bmi, HbA1c_level, blood_glucose_level,
        gender_Female, gender_Male, gender_Other, smoking_history_No_Info,
        smoking_history_current, smoking_history_ever, smoking_history_former,
        smoking_history_never, smoking_history_not_current, health_index,
        log_blood_glucose
    ]
    
    return patient_data

# Get user input and make prediction
user_data = get_user_input()
result = predict_diabetes(user_data)
print(f"The patient is {result}.")
```
